```python
import matplotlib.pyplot as plt
import seaborn as sns
from Environment import MockALGame
import config.mockConfig as config
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.set()
```

```python
env = MockALGame(config)
quality = list()
bvssb = list()
entropy = list()
for _ in range(100):
    state = env.reset()
    quality += list(state[:, 3])
    bvssb += list(state[:, 1])
    entropy += list(state[:, 2])
```

```python
sns.scatterplot(x=quality, y=entropy, alpha=0.7, label="entropy")
sns.scatterplot(x=quality, y=bvssb, alpha=0.7, label="BvsSB")
plt.xlabel("Quality of Sample")
plt.ylabel("Mock Feature")
plt.ylim(-1, 5)
plt.legend()
plt.show()
```
