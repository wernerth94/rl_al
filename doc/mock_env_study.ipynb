```python
import matplotlib.pyplot as plt
import seaborn as sns
from Environment import MockALGame
import config.mockConfig as config
sns.set()
```

```python
env = MockALGame(config)
```

```python
x = list()
bvssb = list()
entropy = list()
for _ in range(50):
    state = env.reset()
    x += list(state[:, 3])
    bvssb += list(state[:, 1])
    entropy += list(state[:, 2])
sns.scatterplot(x=x, y=entropy, alpha=0.7, label="entropy")
sns.scatterplot(x=x, y=bvssb, alpha=0.7, label="BvsSB")
plt.ylim(-1, 5)
plt.legend()
plt.show()
```
